```python
plt.figure(figsize=(8,6))
plt.hist(df['comb'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Combustion')
plt.xlabel('Combustion')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows most vehicles consume between 20–30 L/100km. The distribution is right-skewed, indicating fewer high-consumption vehicles (above 35 L/100km). This pattern highlights the need for improving fuel efficiency in moderately consuming cars, as they make up the majority


```python
plt.figure(figsize=(8,6))
plt.hist(df['fuelType'], bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows that most vehicles use either Premium Gasoline or Regular Gasoline, with significantly fewer vehicles using Diesel, Midgrade Gasoline, or Natural Gas. This distribution is highly right-skewed, indicating that gasoline-powered vehicles dominate the dataset.

```python
# Only numeric columns
numeric_data = df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
corr_matrix = numeric_data.corr()

# Create the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

plt.title("Correlation Heatmap of Vehicle Data")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

The heatmap shows the correlation between all numeric features. Some variables are highly correlated, such as:

- `city` and `UCity` (r ≈ 0.95)
- `highway` and `UHighway` (r ≈ 0.96)

This indicates that `UCity` and `UHighway` are almost duplicates of `city` and `highway`, respectively. We could drop `UCity` and `UHighway` to reduce redundancy and simplify the model.

Also, `fuel_efficiency` is strongly negatively correlated with `co2_emissions`, which is expected: more efficient cars emit less CO2.
